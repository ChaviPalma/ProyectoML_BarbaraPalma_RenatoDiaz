# Fase 1: Business Understanding

KitsuneList es una plataforma de streaming de anime, que busca hacer un apartado de recomendaciones personalizadas para ayudar a sus usuarios que tuvieran dificultades para saber que ver en sus tiempos libres. Las sugerencias que se les daran a los usuarios se generará según los gustos de cada usuario, teniendo en consideración las temáticas, el año de estreno, el género, y su historial de visualizaciones, de esta manera poder facilitar el descubrimiento de nuevos animes y películas.

## Objetivos Principales:

### Desarrollar un Sistema de Recomendaciones Personalizadas



*   Predicción de regresion para la afinidad de los usuarios:


En el contexto del KitsuneList, el objetivo de la predicción que se tiene para estimar la puntuación de la afinidad del anime para cada usuario. Esta puntuación reflejará qué tan relevante podría ser un nuevo anime según el historial de visualizaciones, las calificaciones que obtuvo tal anime y las preferencias de género o temática del usuario. Se prefirió este enfoque por que proporciona un valor cuantitativo que nos permitirá ordenar las recomendaciones de manera más precisa. De este modo, el sistema se puede ajustar las preferencias o sugerencias según intereses individuales, mejorando de esta manera la experiencia de los usuarios y aumentando la probabilidad de disfrute y vea los animes que se le fueron recomendado.



*   Predicción de clasificación del interés del usuario:





El objetivo de esta tarea es poder determinar si un usuario mostrará interés en un anime que se les fue recomendado en la plataforma(respuestas de si o no). Por esto, se consideraran factores como el historial de visualizaciones, géneros preferidos, año de estreno del anime, popularidad del anime y las puntuaciones más altas de animes previamente vistos. Esta clasificación nos permitirá filtrar las recomendaciones más relevantes, así de esta manera se alineará con el objetivo que busca aumentar la retención y el engagement dentro de la plataforma KitsuneList. Ademas, ayudará a priorizar qué animes se mostrarán primero, facilitando el descubrimiento de nuevos animes adaptadas a los gustos de los usuarios.

##¿Qué esperamos lograr?


* Se espera lograr que mediante este sistema de predicción haya un aumento en la retención de los usuarios dentro de la plataforma kitsuneList. También, se espera ser más competitivos en comparación a otras plataformas de streaming de anime, ya que queremos ofrecer una funcionalidad avanzada y personalizada para cada usuario.

* Hay que tener en cuenta que nuestra plataforma de streaming tiene que pagar licencias para cada anime y así poder transmitirlo legalmente. Por lo que es ideal que mediante estos datasets, igual podamos obtener si un anime va a ser popular entre el publico, y así no tener perdidas por animes que no son cosumidos dentro de la plataforma.
